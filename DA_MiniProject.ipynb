```python
# calculate mean of a numeric column
avg_runs = df["batsman_runs"].astype("float").mean(axis=0)
print("Average of batsman_runs:", avg_runs)

# replace NaN with mean
df["batsman_runs"].replace(np.nan, avg_runs, inplace=True)
```

```python
# calculate mean value for "total_runs" column
avg_runs = df["total_runs"].astype("float").mean(axis=0)
print("Average of total_runs:", avg_runs)

# replace NaN by mean value in "total_runs" column
df["total_runs"].replace(np.nan, avg_runs, inplace=True)
```

```python
avg_batsman = df["batsman_runs"].astype("float").mean(axis=0)
print("Average of batsman_runs:", avg_batsman)

df["batsman_runs"].replace(np.nan, avg_batsman, inplace=True)
```

```python
# calculate mean value for "batsman_runs" column
avg_runs = df["batsman_runs"].astype("float").mean(axis=0)
print("Average of batsman_runs:", avg_runs)

# replace NaN by mean value in "batsman_runs" column
df["batsman_runs"].replace(np.nan, avg_runs, inplace=True)
```

```python
avg_extra_runs = df["extra_runs"].astype("float").mean(axis=0)
print("Average of extra_runs:", avg_extra_runs)

df["extra_runs"].replace(np.nan, avg_extra_runs, inplace=True)
```

```python
plt.figure(figsize=(10,6))

# group by team and over → take mean runs
grouped = df.groupby(['batting_team', 'over'])['total_runs'].mean().reset_index()

# plot each team
for team in grouped['batting_team'].unique():
    team_data = grouped[grouped['batting_team'] == team]
    plt.plot(team_data['over'], team_data['total_runs'], marker='o', label=team)

plt.title('Average Runs per Over by Team')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.legend()
plt.show()
```

```python
top_teams = df['batting_team'].value_counts().head(5).index
filtered = df[df['batting_team'].isin(top_teams)]

grouped = filtered.groupby(['batting_team', 'over'])['total_runs'].mean().reset_index()

plt.figure(figsize=(10,6))

for team in top_teams:
    team_data = grouped[grouped['batting_team'] == team]
    plt.plot(team_data['over'], team_data['total_runs'], marker='o', label=team)

plt.title('Top Teams: Avg Runs per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.legend()
plt.show()
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(16,8))

# BEFORE (original data)
plt.subplot(2,2,1)
sns.histplot(df['total_runs'], kde=True)
plt.title('Before - Distribution')

plt.subplot(2,2,2)
sns.boxplot(y=df['total_runs'])
plt.title('Before - Boxplot')

# AFTER (after filling missing values)
new_df = df.copy()
new_df['total_runs'].fillna(new_df['total_runs'].mean(), inplace=True)

plt.subplot(2,2,3)
sns.histplot(new_df['total_runs'], kde=True)
plt.title('After - Distribution')

plt.subplot(2,2,4)
sns.boxplot(y=new_df['total_runs'])
plt.title('After - Boxplot')

plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2,2,1)
sns.histplot(df['batsman_runs'], kde=True)

plt.subplot(2,2,2)
sns.boxplot(y=df['batsman_runs'])

new_df = df.copy()
new_df['batsman_runs'].fillna(new_df['batsman_runs'].mean(), inplace=True)

plt.subplot(2,2,3)
sns.histplot(new_df['batsman_runs'], kde=True)

plt.subplot(2,2,4)
sns.boxplot(y=new_df['batsman_runs'])

plt.show()
```
